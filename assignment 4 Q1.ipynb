```python
page=requests.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
page
```

```python
# Parse the page with BeautifulSoup
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
# Find the table containing the video details
```

```python
table=soup.find('table',{'class':'wikitable sortable'})
```

```python
# Iterate through the rows of the table and extract the desired information
```

```python
for row in table.find_all('tr')[1:31]:
    columns = row.find_all('td')
    rank = columns[0].text.strip()
    name = columns[1].text.strip()
    artist = columns[2].text.strip()
    upload_date = columns[3].text.strip()
    views = columns[4].text.strip()
    print(f"Rank: {rank}")
    print(f"Name: {name}")
    print(f"Artist: {artist}")
    print(f"Upload Date: {upload_date}")
    print(f"Views: {views}")
    
```

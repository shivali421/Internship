```python
#opening naukri page on automated chrome browser
```

```python
driver.get("https://www.naukri.com/datascience-recruiters-consultants")
```

```python
name_desination=[]
```

```python
company=[]
```

```python
skills=[]
```

```python
location=[]
```

```python
start=0
```

```python
end=10
```

```python
#scraping name and desination
```

```python
for page in range(start,end):
    name=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
    for i in name:
        name_desination.append(i.text)
```

```python
#scraping company name
```

```python
for page in range(start,end):
    co=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
    for i in co:
        company.append(i.text)
```

```python
#scraping skills they hire for
```

```python
for page in range(start,end):
    skill=driver.find_elements(By.XPATH,'//ul[@class="tags has-description"]')
    for i in skill:
        skills.append(i.text)
```

```python
#scraping location
```

```python
for page in range(start,end):
    place=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
    for i in place:
        location.append(i.text)
```

```python
next_button=driver.find_element(By.XPATH,'//a[@class="fright fs14 btn-secondary br2"]')
```
